```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Para o acompanhamento do programa segue os dados de diâmetro de cada Planeta
planetas_dataframe =pd.read_excel("planetas.xlsx")
display(planetas_dataframe)
```

```python
Planetas = [planetas_dataframe["Nome"]]
Diâmetros = [planetas_dataframe["Diâmetro"]]

for  planeta in Planetas:
    planeta = planeta


for diâmetro in Diâmetros:
    planeta = planeta


plt.bar(planeta,diâmetro)
plt.title("Diâmetro dos Planetas")
plt.xlabel("Planetas")
plt.ylabel("Diâmetro")
plt.show()
```
